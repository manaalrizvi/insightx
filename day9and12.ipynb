```python
movies = df[df['type'] == 'Movie']
director_counts = movies['director'].value_counts()
```

```python
top=director_counts.value_counts().head(10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plt.pie(top.values,labels=top.index,autopct="%1.1f%%",textprops={'fontsize':8})




plt.title("Top 10 directors")
plt.tight_layout()
plt.show()
```
